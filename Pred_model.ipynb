```python
from torchvision.io import read_image
from torchvision.models import alexnet,AlexNet_Weights
import torchvision.transforms as T
```

# Alexnet

ImageNet ma'lumotlar to'plami - bu WordNet ierarxiyasi bo'yicha tashkil etilgan 14 milliondan ortiq izohli tasvirlardan iborat bo'lgan keng ko'lamli vizual aniqlash muammolari ma'lumotlar to'plami.. Ma'lumotlar to'plami 20 000 dan ortiq turli xil ob'ektlar toifalaridagi tasvirlarni o'z ichiga oladi va uning asosiy maqsadi kompyuterni ko'rish va tasvirni aniqlash algoritmlari sohasidagi tadqiqotlar va taqqoslashlarni osonlashtirishdir.

Ma'lumotlar to'plamidagi har bir rasmga u tasvirlangan ob'ekt toifasiga mos keladigan yorliq bilan izohlanadi. Yorliqlar ierarxik va WordNet taksonomiyasiga muvofiq tashkil etilgan, bu tildagi so'zlar o'rtasidagi munosabatlarni belgilaydigan hisoblash leksikasi. WordNet ierarxiyasi ImageNet toifalarini daraxtga o'xshash tuzilishda tashkil qilish uchun ishlatiladi, daraxtdagi har bir tugun boshqa toifaga mos keladi.

ImageNet ma'lumotlar to'plami turli xil kompyuter ko'rish modellarini, shu jumladan AlexNet modeli va boshqa chuqur o'rganish modellarini o'rgatish va baholash uchun keng qo'llaniladi. Bu chuqur o'rganish texnologiyasini rivojlantirishda hal qiluvchi rol o'ynadi va uning kompyuterni ko'rish sohasiga ta'siri juda katta.

```python
weights = AlexNet_Weights.DEFAULT
model = alexnet(weights=weights)
model.eval()
```

```python
!wget -O cat.jpg "https://daryo.uz/cache/2020/09/photo_2020-09-22_13-51-48-552x414.jpg"
```

```python
img=read_image("cat.jpg")
```

```python
img.shape
```

```python
preprocess=weights.transforms()
pred=preprocess(img).unsqueeze(0)
```

```python
pred.shape
```

```python
r=model(pred)
r.argmax(axis=1)
```

```python
(weights.meta["categories"][281])
```

# EfficientNet

CIFAR-100 dataset tizimida turli xil obyektlarning tasvirlari mavjud. Ushbu dataset 32x32 piksel o'lchamdagi 60,000 ta rangli tasvirlarni o'z ichiga oladi, 100 ta turdagi obyektlar to'plamidan. Har bir obyekt turiga 600 ta tasvir tushirilgan. Shu sababli, bu dataset katta miqdorda Tasvirlarni aniq yoki sinflandirish vazifalarini bajarish uchun ishlatiladi.

 Xususan, bizning EfficientNet-B7 eng yaxshi mavjud ConvNet-ga qaraganda 8,4 baravar kichikroq va 6,1 baravar tezroq xulosa chiqarish bilan birga, ImageNet-da 84,3 foiz yuqori darajadagi eng zamonaviy aniqlikka erishadi. Bizning EfficientNets ham yaxshi uzatadi va CIFAR-100 (91,7%), Gullar (98,8%) va boshqa 3 ta uzatish o'rganish ma'lumotlar to'plamida eng so'nggi aniqlikka erishadi, parametrlari kamroq.

```python
from torchvision.models import efficientnet_b7,EfficientNet_B7_Weights
```

```python
weights=EfficientNet_B7_Weights.DEFAULT
model=efficientnet_b7(weights=weights)
model.eval()
```

```python
preprocess=weights.transforms()
pred=preprocess(img2).unsqueeze(0)
```

```python
c=model(pred)
c.argmax(axis=1)
```

```python
weights.meta['categories'][386]
```
