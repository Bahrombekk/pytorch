```python
import cv2
import matplotlib.pyplot as plt
```

```python
from PIL import Image
import numpy as np

new_size = (224, 224)

def resize_and_pad(img_path):
    img = Image.open(img_path)
    img.thumbnail(new_size, Image.ANTIALIAS)
    width, height = img.size
    x_pad = (new_size[0] - width) // 2
    y_pad = (new_size[1] - height) // 2
    padded_img = Image.new('RGB', new_size, 'black')
    padded_img.paste(img, (x_pad, y_pad))
    img=np.array(padded_img)
    print(img.shape)
    plt.imshow(img)
    return img

# Example usage
img_path = "/content/cat.jpg"
resized_img = resize_and_pad(img_path)
```
