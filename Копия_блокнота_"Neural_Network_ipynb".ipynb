```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from  sklearn.datasets import make_regression
```

#Data 1x1

```python
x,y=make_regression(n_samples=5,n_features=1,n_targets=1,random_state=True)
```

```python
x=torch.tensor(x,dtype=torch.float32)
y=torch.tensor(y,dtype=torch.float32)/100
```

```python
plt.scatter(x,y)
```

```python
model=torch.nn.Linear(in_features=1,out_features=1)
```

```python
critereon=torch.nn.MSELoss()
l=0.01
optimizer=torch.optim.Adam(model.parameters(),lr=l)
```

```python
loss_data=[]
epoch=7
for i in range(epoch):
  critereon.zero_grad()
  y_hat=model(x)
  loss=critereon(y_hat,y)
  loss_data.append(loss.item())
  loss.backward()
  optimizer.step()

  if i%1==0:
    print("i {} , loss {}".format(i,loss.item()))
```

```python
plt.plot(loss_data)
```
